```python
# Save prompt templates as YAML file
prompt_yaml = """
system_prompt:
  prompt: "You are a smart assistant who solves problems step by step using Python tools. Always end with the `final_answer` tool."

final_answer:
  pre_messages: "You have finished the task. Now call `final_answer()` with the result."
  post_messages: "Well done."

planning:
  initial_facts: "### 1. Facts given\\n### 2. Facts to look up\\n### 3. Facts to derive"
  initial_plan: "Plan:\\n- Step 1\\n- Step 2\\n<end_plan>"
  update_facts_pre_messages: "Analyze history and update facts."
  update_facts_post_messages: "Updated facts."
  update_plan_pre_messages: "Build a new plan."
  update_plan_post_messages: "Updated plan:\\n<end_plan>"

managed_agent:
  task: "You are agent '{{name}}'. Your task is:\\n{{task}}"
  report: "Agent '{{name}}' says:\\n{{final_answer}}"
"""

# Write it to file
with open("prompts.yaml", "w") as f:
    f.write(prompt_yaml)
```

```python
import yaml

with open("prompts.yaml", "r") as f:
    prompt_templates = yaml.safe_load(f)
```

```python
from smolagents import CodeAgent
import wikipedia
from serpapi import GoogleSearch
from smolagents import tool

@tool
def serp_search(query: str) -> str:
    """Searches the web using SerpApi and returns the top result title and link.

    Args:
        query (str): The search query string.
    """
    params = {
        "q": query,
        "api_key": "3af865b088184a15f55466853d796e1dee3735b97118a22d108c7a2e4b16d8c8"  # <-- Replace this with your real API key
    }
    search = GoogleSearch(params)
    results = search.get_dict()

    try:
        top_result = results["organic_results"][0]
        return f"{top_result['title']} - {top_result['link']}"
    except (KeyError, IndexError):
        return "No results found."
@tool
def wikipedia_search(query: str) -> str:
    """
    Fetches a short summary from Wikipedia for the given topic.

    Args:
        query (str): The topic to search on Wikipedia.

    Returns:
        str: A short summary about the topic.
    """
    try:
        return wikipedia.summary(query, sentences=2)
    except Exception as e:
        return str(e)

agent = CodeAgent(
    tools=[wikipedia_search,serp_search],
    model=model,
    max_steps=4,
    verbosity_level=2,
    add_base_tools=False

    )

print(agent.tools.keys())
response = agent.run("Using only the provided serp_search search and wikipedia_search, what is 2+2, relying on the search results.")
print(response)
```
